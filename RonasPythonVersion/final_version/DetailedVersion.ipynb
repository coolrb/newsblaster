# DataPrep

Dataprep calculates the tf-idfs of the given set of documents and saves them to a file
For more information on tf-idfs: http://en.wikipedia.org/wiki/Tf%E2%80%93idf
For more information on the specifics of this tf-idfs calculation: http://scikit-learn.org/stable/modules/generated/sklearn.feature_extraction.text.TfidfVectorizer.html

```python
from dataprep import dataprep
import os
import math
import numpy as np
import codecs
from sklearn.feature_extraction.text import TfidfVectorizer
import json
```

Parameters of DataPrep

```python
root_path = '../pres_root_dir_1/' #Location of Root Directory containing a documents folder with the documents
save_path = '../pres_root_dir_1/save_tfidfs' #Location to save the resulting tf-idf
```

Get file with stopwords and split by words. (Stop words are words ignored in the tf-idfs counts).

```python
stop_words_path = 'stoplist'
stop_words_file = open(stop_words_path, 'r')
stop_words_list = stop_words_file.read().splitlines()
```

Get the document text and the names of the documents and save them into two lists with the same indicies for each document. Note that the documents are located at [root_path]/documents

```python
list_of_docs = []
list_doc_ids = []
documents_folder = 'documents'
documents_path = os.path.join(root_path, documents_folder)
documents_in_folder = os.listdir(documents_path)
documents_in_folder = sorted(documents_in_folder)
for document in documents_in_folder:
    with codecs.open(os.path.join(documents_path, document),
                     'r', 'utf-8', errors='replace') as f:
        document_text = u''.join(f.readlines())
        list_of_docs.append(document_text)
        list_doc_ids.append(document)
```

```python
print list_doc_ids
```

Use SkLearn's module to calculate tf-idf

```python
tfidf = TfidfVectorizer(stop_words=stop_words_list,
                        norm=None, smooth_idf=False)
tfidfs = tfidf.fit_transform(list_of_docs)
```

Write the information to the given file. Each document is represented as a dictionary containing the key:value pairs of word:tf-idf.
Each document is then placed into a list in the same order as the list_doc_ids. Both the list of dictionaries and the list_doc_ids are then dumped into the specified save_path file.

Create the individual dictionaries for each document

```python
list_of_dicts = []
transformed_matrix = tfidf.inverse_transform(tfidfs)
feature_names = tfidf.get_feature_names()
for i in range(0, len(transformed_matrix)):
    article = transformed_matrix[i]
    article_dict = {}
    for word in article:
        word_index = feature_names.index(word)
        article_dict[word] = tfidfs[i, word_index]
    list_of_dicts.append(article_dict)
```

Dump all information into save_path

```python
full_list = [list_of_dicts, list_doc_ids]
with open(save_path, 'w') as outfile:
    json.dump(full_list, outfile)
```

# Clustering

```python
from clustering import clustering
from clust import clust
import numpy as np
import operator
import sys
import os
import math
from sklearn.metrics import pairwise_distances
from sklearn.feature_extraction import DictVectorizer
from scipy.spatial.distance import cosine
import heapq
from heapq import heappush, heappop
import scipy.sparse
import random
import json
```

We will now cluster this set of documents into super and subclusters. 
The relevant parameters are as follows.

```python
list_file = save_path #Location where the tf-idfs were saved
major_threshold = 0.2 #Used to create superclusters
minor_threshold = 0.5 #Used to create subclusters
```

Load the tf-idfs from the saved file.

```python
list_of_dicts = []
list_doc_ids = []
with open(list_file, 'r') as infile:
    full_list = json.load(infile)
    list_of_dicts = full_list[0]
    list_doc_ids = full_list[1]
```

Get only the first 101 largest tf-idfs from each of the documents

```python
for i in range(0, len(list_of_dicts)):
    dic = list_of_dicts[i]
    sorted_dic = sorted(dic.items(), key=operator.itemgetter(1))
    j = 0
    new_dic = {}
    while j < 101 and j < len(sorted_dic):
        new_dic[sorted_dic[len(sorted_dic) - j - 1][0]] = \
            sorted_dic[len(sorted_dic) - j - 1][1]
        j += 1
    list_of_dicts[i] = new_dic
```

Get the cosine distances between documents based on the tf-idfs.

```python
dict_vect = DictVectorizer()
tfidfs = dict_vect.fit_transform(list_of_dicts)
cosine_distance = 1-pairwise_distances(tfidfs, metric="cosine")
```

Calculate the document distances based on the cosine distances
Each distance between documents is calculated as follows:

(cosine_dis_value is the cosine distance between the documents)

```python
def compute_doc_dis(cosine_dis_value):
    # Get document distance
    h = -3.247626 + 23.595831 * cosine_dis_value
    ex = math.e**h
    return ex / (1.0 + ex)
```

Calculate the document distance between all documents

```python
f = np.vectorize(compute_doc_dis)
docs_dist = f(cosine_distance) 
```

Make each document into a cluster. We will merge these clusters as we go along. 

```python
list_of_clusts = []
for i in range(0, len(list_doc_ids)):
    initialClust = clust([i], i)
    list_of_clusts.append(initialClust)
list_of_clusts = np.array(list_of_clusts)
```

Create an array with the distances between each cluster. 

```python
clust_dist = np.zeros([list_of_clusts.shape[0],
                               list_of_clusts.shape[0]])
```

Each individual distance between two clusters is calculated as follows.

```python
def compute_new_clust_dis(docs_dist, clust1, clust2):
    # Average of document pair distances
    # Where each pair must have one doc from clust1 and one from clust2
    num_docs = 0
    tot_docs_dist = 0
    for doc1 in clust1.docs:
        for doc2 in clust2.docs:
            num_docs += 1
            tot_docs_dist += docs_dist[doc1, doc2]
    return tot_docs_dist / num_docs
```

To save the actual distances into the array and to avoid repeated calculations there is the following function. 

```python
def compute_clust_dis(clust_dist, docs_dist, clust1, clust2):
    # Make clust1 the smaller of the two clust ids
    if clust1.clust_id > clust2.clust_id:
        temp = clust1
        clust1 = clust2
        clust2 = temp
    # Retrieve clust dist if already calculated, otherwise recalculate
    if (clust_dist[clust1.clust_id, clust2.clust_id] == 0):
        clust_dist[clust1.clust_id, clust2.clust_id] =  \
            compute_new_clust_dis(docs_dist, clust1, clust2)
    return clust_dist[clust1.clust_id, clust2.clust_id]
```

Update the array that we set aside for the distances between clusters. 

```python
for i in range(0, len(list_of_clusts)):
            for j in range(i + 1, len(list_of_clusts)):
                clust_dist[list_of_clusts[i].clust_id,
                           list_of_clusts[j].clust_id] = \
                    compute_clust_dis(clust_dist, docs_dist,
                                           list_of_clusts[i],
                                           list_of_clusts[j])
print clust_dist
```

Go through all the cluster pairs and find any pairs whose distance meets the major threshold requirement. If the requirement is met, then add the cluster to a heap (ordered by most similar to list similar clusters - i.e. by cluster similarity). 

Merge the pairs of clusters that are on the heap. After each merge, remove the pairs that no longer exist (because the merge removed one of the two clusters in the pair). Then go through all the cluster pairs again and see if any new cluster pairs now meet the threshold. Repeat while there are still cluster pairs on the heap. 

```python
def merge_cluster_set(clusts_to_merge, docs_dist,
                      clust_dist, threshold):
    # Find all clust pairs that exceed the threshold req to merge
    # and add them to a heap (sorted by clust similarity)
    merge_pairs = []
    for i in range(0, len(clusts_to_merge)):
        for j in range(i + 1, len(clusts_to_merge)):
            curr_dist = compute_clust_dis(clust_dist, docs_dist,
                                               clusts_to_merge[i],
                                               clusts_to_merge[j])
            if curr_dist > threshold:
                heappush(merge_pairs, (1-curr_dist,
                                       clusts_to_merge[i],
                                       clusts_to_merge[j]))
    # Merge the clust pairs that are now on the heap
    while merge_pairs:
        pair = heappop(merge_pairs)
        clust1 = pair[1]
        clust2 = pair[2]
        merged_result = merge_clust(clusts_to_merge, clust1, clust2,
                                         docs_dist, clust_dist)
        clust1 = merged_result[0]
        clusts_to_merge = merged_result[1]
        clust_dist = merged_result[2]
        # Remove pairs on heap that no longer exist because of the merge
        merge_pairs = remove_affected_pairs(merge_pairs,
                                                 clust1, clust2)
        # Calculate clust distances to this new heap
        for my_clust in clusts_to_merge:
            if my_clust.clust_id != clust1.clust_id:
                curr_dist = compute_clust_dis(clust_dist, docs_dist,
                                                   clust1, my_clust)
                # If the new clust dist meets the threshold, add to heap
                if curr_dist > threshold:
                    if clust1.clust_id > my_clust.clust_id:
                        heappush(merge_pairs, (1-curr_dist, my_clust, clust1))
                    else:
                        heappush(merge_pairs, (1-curr_dist, clust1, my_clust))
    return clusts_to_merge, clust_dist
```

Call the merge function with the correct parameters and save the results. 

```python
result = merge_cluster_set(list_of_clusts, docs_dist,
                                            clust_dist, major_threshold)
existing_clust = result[0]
clusts_dist = result[1]
```

Merge the pairs of clusters that are on the heap. Each merge is performed as follows. 

```python
def merge_clust(list_of_clusts, clust1, clust2,
                    docs_dist, clust_dist):
        # Compute clust distances to this new pair
        for my_clust in list_of_clusts:
            if my_clust.clust_id != clust1.clust_id and \
               my_clust.clust_id != clust2.clust_id:
                dist_to_clust1 = compute_clust_dis(clust_dist, docs_dist,
                                                        my_clust, clust1)
                dist_to_clust2 = compute_clust_dis(clust_dist, docs_dist,
                                                        my_clust, clust2)
                if my_clust.clust_id < clust1.clust_id:
                    clust_dist[my_clust.clust_id, clust1.clust_id] = \
                        merged_clust_dis(dist_to_clust1, len(clust1.docs),
                                              dist_to_clust2, len(clust2.docs))
                else:
                    clust_dist[clust1.clust_id, my_clust.clust_id] = \
                        merged_clust_dis(dist_to_clust1,
                                              len(clust1.docs), dist_to_clust2,
                                              len(clust2.docs))
        # Combine the documents and subclusters of the two clusts
        clust1.docs = np.append(clust1.docs, clust2.docs)
        clust1.subclusts = np.append(clust1.subclusts, clust2.subclusts)
        # Update the new clust in list of clusts
        for this_clust in list_of_clusts:
            if this_clust.clust_id == clust1.clust_id:
                my_clust = clust1
                break
        # Remove the second clust
        list_of_clusts = delete_clust(list_of_clusts, clust2)
        return clust1, list_of_clusts, clust_dist
```

Auxiliary Functions for Merging.
First function calculates the distance between clusters after the merge occurs.
Section function deletes the cluster (from the pair of clusters that was merged) with the larger id.

```python
def merged_clust_dis(dist_to_clust1, num_docs_clust1, dist_to_clust2,
                     num_docs_clust2):
    # Calculate merge distance
    return (dist_to_clust1 * num_docs_clust1 + dist_to_clust2 *
            num_docs_clust2) / (num_docs_clust1 + num_docs_clust2)

def delete_clust(list_of_clusts, del_clust):
    # Deletes a clust from the given list based on clust_id
    modified_list = []
    for my_clust in list_of_clusts:
        if my_clust.clust_id != del_clust.clust_id:
            modified_list.append(my_clust)
    return np.array(modified_list)
```

To remove cluster pairs that no longer exists after the merge is complete:

```python
def remove_affected_pairs(merge_pairs, clust1, clust2):
    # Makes sure the pairs in merge pairs do no contain clust1 or clust2
    # If it does, remove the pair containing either clust1 or clust2
    result_heap = []
    while merge_pairs:
        pair = heappop(merge_pairs)
        if clust1.clust_id != pair[1].clust_id and \
           clust1.clust_id != pair[2].clust_id and \
           clust2.clust_id != pair[1].clust_id and \
           clust2.clust_id != pair[2].clust_id:
            heappush(result_heap, pair)
    return result_heap
```

The superclusters are now formed! To form the subclusters we repeat the same process above with the heap, merging, updating heap process with the minor threshold instead of the major threshold. 

So we can call this function in a loop to apply to each of the super clusters. 

```python
for superclust in list_of_clusts:
    clusts_dist = np.zeros([len(superclust.docs),
                            len(superclust.docs)])
    list_of_subclusts = []
    next_clust_id = 0
    for doc in superclust.docs:
        subclust = clust([doc], next_clust_id)
        list_of_subclusts.append(subclust)
        next_clust_id += 1
    list_of_subclusts = np.array(list_of_subclusts)
    subclusts = merge_cluster_set(list_of_subclusts, docs_dist,
                                       clusts_dist,
                                       minor_threshold)
    superclust.subclusts = subclusts[0]
```
